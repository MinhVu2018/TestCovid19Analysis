```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

data source: 
https://ourworldindata.org/coronavirus/country/vietnam

```python
data = pd.read_csv(r"owid-covid-data.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.location.unique()
```

```python
VN_data = data.loc[data['location'] == 'Vietnam']
VN_data
```

```python
# VN_data.dropna(axis=1)
VN_data = VN_data.dropna(axis=1, how='all')
```

```python
VN_data.columns
```

```python
VN_data = VN_data[['date', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'total_tests', 'new_tests', 'total_vaccinations', 'new_vaccinations', 'population']]
VN_data.fillna(0)
```

```python
sns.relplot(x='date', y='total_cases', data = VN_data)
```

```python
sns.pairplot(VN_data)
```

```python
sns.relplot(x='date', y='total_deaths', kind='line', data = VN_data)
```

```python
sns.catplot(x='date', y='total_vaccinations', data = VN_data)
```

```python
df1 = VN_data[['new_cases']]
df2 = VN_data[['new_deaths']]
ax = df1.plot()

df2.plot(ax=ax)
```
